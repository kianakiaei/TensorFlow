```python
import tensorflow as tf
import numpy as np
```

```python
# Create a vector, scalar, matrix and tensor with values of your choosing

scalar = tf.constant([1])
vector = tf.constant([1, 2])
matrix = tf.constant([[1, 2],
                      [3, 4],
                      [5, 6]])
tensor = tf.constant([[[1, 2],
                       [3, 4]],
                      [[5, 6],
                       [7, 8]]])
```

```python
print("scalar is:", scalar, "\n")
print("vector is:", vector, "\n")
print("matrix is:", matrix, "\n")
print("tensor is:", tensor, "\n")
```

```python
# Find the shape, rank and size of the tensors you created

print("Shape of tensor is:", tensor.shape)
print("rank of tensor is:", tensor.ndim)
print("size of tensor is:", tf.size(tensor))
```
